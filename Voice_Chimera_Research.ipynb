```python
def lerp(em1, em2, alpha):
    return (1 - alpha) * em1 + alpha * em2
```

```python
import torch
import numpy as np
from typing import Union

def slerp(
    embedding1: Union[torch.Tensor, np.ndarray],
    embedding2: Union[torch.Tensor, np.ndarray],
    t: float
) -> torch.Tensor:
    # Ensure embeddings are PyTorch tensors
    emb1 = embedding1 if isinstance(embedding1, torch.Tensor) else torch.tensor(embedding1, dtype=torch.float32)
    emb2 = embedding2 if isinstance(embedding2, torch.Tensor) else torch.tensor(embedding2, dtype=torch.float32)

    if emb1.shape != emb2.shape:
        raise ValueError("Embeddings must have the same shape")

    # Normalize the embeddings
    emb1 = emb1 / torch.norm(emb1)
    emb2 = emb2 / torch.norm(emb2)

    # Compute the cosine of the angle between the embeddings
    dot = torch.sum(emb1 * emb2)
    dot = torch.clamp(dot, -1.0, 1.0)  # Numerical stability

    # Compute the angle (omega)
    omega = torch.acos(dot)
    sin_omega = torch.sin(omega)

    if sin_omega == 0:
        # If angle is 0 or pi, fall back to linear interpolation
        return (1.0 - t) * emb1 + t * emb2

    # Perform Slerp
    s1 = torch.sin((1.0 - t) * omega) / sin_omega
    s2 = torch.sin(t * omega) / sin_omega
    interpolated = s1 * emb1 + s2 * emb2

    # Normalize the result
    interpolated = interpolated / torch.norm(interpolated)

    return interpolated
```

```python
import numpy as np
from typing import Union

def genetic_combination(
    embedding1: Union[torch.Tensor, np.ndarray],
    embedding2: Union[torch.Tensor, np.ndarray],
    crossover_points: int = 5,
    mutation_rate: float = 0.01,
    mutation_strength: float = 0.1
) -> torch.Tensor:
    # Ensure embeddings are PyTorch tensors
    emb1 = embedding1 if isinstance(embedding1, torch.Tensor) else torch.tensor(embedding1, dtype=torch.float32)
    emb2 = embedding2 if isinstance(embedding2, torch.Tensor) else torch.tensor(embedding2, dtype=torch.float32)

    if emb1.shape != emb2.shape:
        raise ValueError("Embeddings must have the same shape")

    # Generate crossover points
    crossover_indices = sorted(np.random.choice(range(1, emb1.shape[0]), crossover_points, replace=False))

    # Initialize the new embedding
    new_embedding = torch.zeros_like(emb1)

    # Perform crossover
    current_parent = np.random.choice([0, 1])  # Randomly choose starting parent
    start = 0
    for end in crossover_indices:
        new_embedding[start:end] = emb1[start:end] if current_parent == 0 else emb2[start:end]
        current_parent = 1 - current_parent  # Switch parent
        start = end

    # Fill in the last segment
    new_embedding[start:] = emb1[start:] if current_parent == 0 else emb2[start:]

    # Apply mutations
    mutations = torch.rand_like(new_embedding) < mutation_rate
    mutation_values = torch.randn_like(new_embedding[mutations])
    new_embedding[mutations] += mutation_values * mutation_strength * new_embedding[mutations].abs()

    # Normalize the new embedding
    new_embedding = new_embedding / torch.norm(new_embedding)

    return new_embedding
```

```python
import numpy as np
from sklearn.decomposition import PCA

def pca_speaker_mixing(all_speaker_embeddings, embedding1, embedding2, n_components=None):
    # Ensure all inputs are numpy arrays
    if torch.is_tensor(all_speaker_embeddings):
        all_speaker_embeddings = all_speaker_embeddings.cpu().numpy()
    if torch.is_tensor(embedding1):
        embedding1 = embedding1.cpu().numpy()
    if torch.is_tensor(embedding2):
        embedding2 = embedding2.cpu().numpy()

    # Determine the maximum number of components
    max_components = min(all_speaker_embeddings.shape)
    if n_components is None or n_components > max_components:
        n_components = max_components

    # Fit PCA on all speaker embeddings
    pca = PCA(n_components=n_components)
    pca.fit(all_speaker_embeddings)

    # Transform the specific speaker embeddings we want to mix
    embeddings_to_mix = np.vstack([embedding1, embedding2])
    transformed_embeddings = pca.transform(embeddings_to_mix)

    # Mix in the transformed space (simple average)
    mixed_embedding = np.mean(transformed_embeddings, axis=0)

    # Inverse transform to get back to the original space
    mixed_embedding_original = pca.inverse_transform(mixed_embedding.reshape(1, -1))

    mixed_embedding_tensor = torch.tensor(mixed_embedding_original.flatten(), dtype=torch.float32)

    return mixed_embedding_tensor
```

```python
# this function allows to append the new embedding to the existing embedding tensor.
def append_embedding(embedding_table, new_embedding):
    new_emb_table = F.pad(embedding_table, (0, 0, 0, 1), "constant", 0)
    new_emb_table[-1] = new_embedding
    return new_emb_table

# this function deletes the last embedding in the embedding table
def delete_last_embedding(embedding_table):
    embedding_table = embedding_table[:-1]

    return embedding_table

# this function generates the audio with corresoponding speaker embedding and txt
def generate_audio(net_g_ms, sid, stn_tst):
  sid = torch.LongTensor([sid])
  with torch.no_grad():
    x_tst = stn_tst.unsqueeze(0)
    x_tst_lengths = torch.LongTensor([stn_tst.size(0)])
    audio = net_g_ms.infer(x_tst, x_tst_lengths, sid=sid, noise_scale=.667, noise_scale_w=0.8, length_scale=1)[0][0,0].data.float().numpy()
  return audio

def display_audio(audio):
  ipd.display(ipd.Audio(audio, rate=hps_ms.data.sampling_rate))
```

```python
import numpy as np
import librosa

def mcd(audio1, audio2, sr=22050, n_mfcc=13):
    """
    Calculate Mel Cepstral Distortion between two audio signals.

    :param audio1: First audio array
    :param audio2: Second audio array
    :param sr: Sampling rate
    :param n_mfcc: Number of MFCCs to use
    :return: MCD value
    """
    # Ensure same length
    min_len = min(len(audio1), len(audio2))
    audio1, audio2 = audio1[:min_len], audio2[:min_len]

    # Calculate MFCCs
    mfcc1 = librosa.feature.mfcc(y=audio1, sr=sr, n_mfcc=n_mfcc)
    mfcc2 = librosa.feature.mfcc(y=audio2, sr=sr, n_mfcc=n_mfcc)

    # Calculate MCD
    diff = mfcc1 - mfcc2
    mcd = np.sqrt(2 * np.sum(diff**2, axis=0))
    mcd = np.mean(mcd)  # Average over all frames

    return (10 / np.log(10)) * mcd

def euclidean_distance(emb1, emb2):
    """
    Calculate Euclidean distance between two embeddings.

    :param emb1: First embedding
    :param emb2: Second embedding
    :return: Euclidean distance
    """
    return np.linalg.norm(emb1 - emb2)

def cosine_similarity(emb1, emb2):
    """
    Calculate cosine similarity between two embeddings.

    :param emb1: First embedding
    :param emb2: Second embedding
    :return: Cosine similarity
    """
    return np.dot(emb1, emb2) / (np.linalg.norm(emb1) * np.linalg.norm(emb2))
```

```python
import os
import soundfile as sf
# Store the original embedding space (only the first 109 embeddings)

def run_test(test_id, em_id1, em_id2, method, **params):

    stn_tst = get_text("We propose VITS, Conditional Variational Autoencoder with Adversarial Learning for End-to-End Text-to-Speech.", hps_ms)

    em1 = net_g_ms.emb_g.weight.data[em_id1]
    em2 = net_g_ms.emb_g.weight.data[em_id2]

    if method == 'lerp':
        blended_emb = lerp(em1, em2, params['alpha'])
    elif method == 'slerp':
        blended_emb = slerp(em1, em2, params['alpha'])
    elif method == 'genetic':
        blended_emb = genetic_combination(em1, em2, params['crossover_pts'], params['mutation_rate'], params['mutation_str'])
    elif method == 'pca':
        blended_emb = pca_speaker_mixing(net_g_ms.emb_g.weight.data, em1, em2, params['n_components'])

    # Add the new embedding to the model
    new_sid = len(net_g_ms.emb_g.weight.data)
    net_g_ms.emb_g.weight.data = append_embedding(net_g_ms.emb_g.weight.data, blended_emb)

    # Generate audio
    audio_blended = generate_audio(net_g_ms, new_sid, stn_tst)
    audio1 = generate_audio(net_g_ms, em_id1, stn_tst)
    audio2 = generate_audio(net_g_ms, em_id2, stn_tst)

    # Save audio files
    os.makedirs('generated_audio', exist_ok=True)
    sf.write(f'generated_audio/test_{test_id}_blended.wav', audio_blended, hps_ms.data.sampling_rate)
    sf.write(f'generated_audio/test_{test_id}_speaker1.wav', audio1, hps_ms.data.sampling_rate)
    sf.write(f'generated_audio/test_{test_id}_speaker2.wav', audio2, hps_ms.data.sampling_rate)

    # Calculate metrics
    cos_sim1 = cosine_similarity(blended_emb.cpu().numpy(), em1.cpu().numpy())
    cos_sim2 = cosine_similarity(blended_emb.cpu().numpy(), em2.cpu().numpy())
    euc_dist1 = euclidean_distance(blended_emb.cpu().numpy(), em1.cpu().numpy())
    euc_dist2 = euclidean_distance(blended_emb.cpu().numpy(), em2.cpu().numpy())
    mcd1 = mcd(audio_blended, audio1, sr=hps_ms.data.sampling_rate)
    mcd2 = mcd(audio_blended, audio2, sr=hps_ms.data.sampling_rate)

    return {
        'test_id': test_id,
        'new_sid': new_sid,
        'cosine_similarity1': cos_sim1,
        'cosine_similarity2': cos_sim2,
        'euclidean_distance1': euc_dist1,
        'euclidean_distance2': euc_dist2,
        'mcd1': mcd1,
        'mcd2': mcd2,
        'audio_path': f'generated_audio/test_{test_id}_blended.wav'
    }
```

```python
# Get all speaker embeddings
# Run all tests
results = []

# Lerp and Slerp
results.append(run_test(1, 4, 100, 'lerp', alpha=0.5))
results.append(run_test(2, 4, 100, 'slerp', alpha=0.5))

# Genetic
for test_id in range(3, 30):
    crossover_pts = 2 if test_id <= 11 else (5 if test_id <= 20 else 25)
    mutation_rate = 0.01 if test_id in [3,4,5,12,13,14,21,22,23] else (0.1 if test_id in [6,7,8,15,16,17,24,25,26] else 0.2)
    mutation_str = 0.05 if test_id in [3,6,9,12,15,18,21,24,27] else (0.1 if test_id in [4,7,10,13,16,19,22,25,28] else 0.2)
    results.append(run_test(test_id, 4, 100, 'genetic', crossover_pts=crossover_pts, mutation_rate=mutation_rate, mutation_str=mutation_str))

# PCA
for test_id, n_components in zip(range(30, 34), [16, 32, 64, 128]):
    results.append(run_test(test_id, 4, 100, 'pca', n_components=n_components))

# Print or save results
for result in results:
    print(f"Test ID: {result['test_id']}, New SID: {result['new_sid']}")
    print(f"Cosine Similarity: {result['cosine_similarity1']}, {result['cosine_similarity2']}")
    print(f"Euclidean Distance: {result['euclidean_distance1']}, {result['euclidean_distance2']}")
    print(f"MCD: {result['mcd1']}, {result['mcd2']}")
    print(f"Audio file: {result['audio_path']}")
    print("---")
```
